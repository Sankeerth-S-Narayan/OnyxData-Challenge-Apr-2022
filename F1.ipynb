### Treating missing values

#### Replacing the missing values in column 'code' with the first three characters of surname

```python
df['code']=df['Driver\'s surname'].str[:3]
```

```python
df['code']=df['code'].str.upper()
```

```python
df.position
```

#### The 'position' column can be dropped as both 'position' and 'positionOrder' columns are the same

```python
df.drop(['position'],axis=1,inplace=True)
```

#### The missing values in 'fastestLapSpeed', 'fastestLapTime', 'fastestLap' is because the drivers did not complete the race due to various reasons. So we will replace these values with 0 which means they did not complete the race

```python
df['fastestLapSpeed'].fillna(value=0,axis=0,inplace=True)
df['fastestLapTime'].fillna(value=0,axis=0,inplace=True)
df['fastestLap'].fillna(value=0,axis=0,inplace=True)
```

```python
df['rank'].fillna(value=0,axis=0,inplace=True)
```

```python
df.drop(['time'],axis=1,inplace=True)
```

```python
df['milliseconds'].fillna(value=0,axis=0,inplace=True)
```

```python
df.isnull().sum()
```

```python
df[["Country","Event"]] = df['Event name'].str.split(" ", n=1,expand = True)
```

#### Converting the treated dataset into csv

```python
df.to_csv('f1.csv', encoding='utf-8')
```
